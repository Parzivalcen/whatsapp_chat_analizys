```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import emoji
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import warnings
warnings.filterwarnings("ignore")

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
print('Packages imported...')
```

```python
import regex
def date_time(s):
    pattern = '^([0-9]+)(\/)([0-9]+)(\/)([0-9]+), ([0-9]+):([0-9]+)[ ]? -'
    result = regex.match(pattern, s)
    if result:
        return True
    return False

def find_author(s):
    s = s.split(': ')
    if len(s)==2:
        return True
    else:
        return False

def getDatapoint(line):
    splitline = line.split(' - ')
    dateTime = splitline[0]
    date, time = dateTime.split(', ')
#     take off the brakets
    message = ''.join(splitline[1:])
    if find_author(message):
        splitmessage = message.split(': ')
        author = splitmessage[0]
        message = ''.join(splitmessage[1:])
    else:
        author = None
    return date, time, author, message
```

```python
data = []
conversation = 'WhatsApp Chat with Adal.txt'
with open(conversation, encoding='utf-8') as fp:
    fp.readline()
    messageBuffer = []
    date, time, author = None, None, None
    while True:
        line = fp.readline()
        if not line:
            break
        # Remove space at beggining and at the end of a string.
        line = line.strip()
        if date_time(line):
            if len(messageBuffer) > 0:
                data.append([date, time, author,' '.join(messageBuffer)])
            # clean the messageBuffer list every time the loop runs. 
            messageBuffer.clear()
            date, time, author, message = getDatapoint(line)
            messageBuffer.append(message)
        else:
            messageBuffer.append(line)
```

```python
df = pd.DataFrame(data, columns=['Date', 'Time', 'Author', 'Message'])
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
print('Delete rows containing "<Media omitted>" ',df[df['Message']=='<Media omitted>'].shape[0])
df.drop(df[df['Message']=='<Media omitted>'].index, inplace = True)
```

- Exracting emojis, but I do not thing there is any

```python
def split_count(text):
    emoji_list = []
#     make a list with every character 
    data = regex.findall(r'\X',text)
    for word in data:
        if any(char in emoji.UNICODE_EMOJI for char in word):
            emoji_list.append(word)
    return emoji_list

df['emoji'] = df["Message"].apply(split_count)

print('total emojis',sum(df['emoji'].str.len()))
df.head()
```

- Lets check for URL's

```python
urlpattern = r'(https?://\S+)'
df['url_count'] = df.Message.apply(lambda x: regex.findall(urlpattern, x)).str.len()
links = np.sum(df.Message.apply(lambda x: regex.findall(urlpattern, x)).str.len())
df['Shared_link'] = df.Message.apply(lambda x: regex.findall(urlpattern, x))
df['Shared_link'] = df['Shared_link'].str[0]
print(links)
df
```

```python
messages_df = df
messages_df['Characters_count'] = messages_df['Message'].apply(lambda s: len(s))
messages_df['Word_count'] = messages_df['Message'].apply(lambda s: len(s.split(' ')))
messages_df['Message_count'] = 1
messages_df
```

```python
l = ['David Centurión', 'Adal']
for i in range(len(l)):
    # filter the message of the users in each loop
    req_df=messages_df[messages_df['Author']==l[i]]
    # req_df contains the messages of only one user at a time
    print(f'Stats of {l[i]}')
    
    # shape gives the number of rows and columns  
    print(f'Messages sent {req_df.shape[0]}')
    
    avg_words = (np.sum(req_df.Word_count)) / len(req_df)
    print('Average words per message',avg_words)
    
    links = sum(req_df['url_count'])
    print('links shared', links)
    print('')
```
